**Introduction**

In the world of retail, understanding consumer purchasing behavior is critical for optimizing business strategies. This project focuses on analyzing grocery transactions using **Association Rule Mining** and the **Apriori Algorithm**. By applying these techniques to transaction data, we aim to identify patterns in consumer purchases, allowing for targeted product recommendations and more informed business decisions.

We begin by preprocessing and cleaning the dataset of grocery transactions, followed by an **Exploratory Data Analysis (EDA)** to understand the purchasing habits of customers. Using the **Apriori Algorithm**, we identify **frequent itemsetsâ€”sets** of products that are often bought together. From these itemsets, we derive **association rules** that reveal relationships between products, helping us understand how certain items tend to be purchased in combination.

By the end of this project, we will have extracted actionable insights that can be used to recommend products, optimize store layouts, and design promotional strategies that align with customer preferences.

## <a id="2-data-loading-and-preprocessing"></a>**2. Data Loading and Preprocessing**


##### <a id="2-1-importing-required-libraries"></a>**2.1 Importing Required Libraries**
We import necessary libraries for data handling, visualization, and market basket analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import networkx as nx
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Ensure mlxtend is installed
#!pip install mlxtend (In my case I already have mlxtend installed, if you don't then simply uncomment this row and install it)
```

##### <a id="2-2-loading-the-dataset"></a>**2.2 Loading the Dataset**

```python
df = pd.read_csv("Groceries_dataset.csv")
df.head()
```

##### <a id="2-3-checking-for-missing-values-and-duplicates"></a>**2.3 Checking for Missing Values and Duplicates**
Before proceeding, we check for missing values and duplicate records.

```python
print(df.isnull().sum())  # Check for missing values
```

```python
print(f"Duplicate rows: {df.duplicated().sum()}")  # Check for duplicates
```

As we can see here our dataset contains 759 duplicated rows, now let's try to drop them by using <code>drop_duplicates</code> function.

```python
df.drop_duplicates(inplace=True)
print(f"Duplicate rows after removal: {df.duplicated().sum()}")
```

##### <a id="2-4-creating-a-unique-transaction-id"></a>**2.4 Creating a Unique Transaction ID**
To analyze transactions effectively, we combine **Member_number** and **Date** columns to create a **TransactionID**, this allows us to track purchases at a transaction level.

```python
df['Member_number'] = df['Member_number'].astype(str)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['TransactionID'] = df['Member_number'] + '-' + df['Date'].dt.strftime('%Y-%m-%d')
```

##### <a id="2-5-standardizing-product-names"></a>**2.5 Standardizing Product Names**
To ensure consistency and avoid duplicate entries due to case differences, we convert product names to lowercase. This step helps in avoiding discrepancies in item names.

```python
df["itemDescription"] = df["itemDescription"].str.lower()
```

## <a id="4-apriori-algorithm"></a>**4. Apriori Algorithm**

The Apriori Algorithm is a classical algorithm used in association rule mining, which is mainly used to find frequent itemsets in a transaction database. This algorithm identifies the relationships between items in a large dataset by applying the support and confidence metrics to generate association rules.




##### <a id="4-1-identify-frequent-itemsets"></a>**4.1 Identify Frequent Itemsets**

```python
# Group by TransactionID and create a list of items
transactions = df.groupby('TransactionID')['itemDescription'].apply(list).reset_index(name='Items')

# Use TransactionEncoder
te = TransactionEncoder()
te_array = te.fit_transform(transactions['Items'])

# Convert to DataFrame
basket = pd.DataFrame(te_array, columns=te.columns_)

# Add TransactionID back to the DataFrame
basket['TransactionID'] = transactions['TransactionID']
# Drop the TransactionID column before applying Apriori
basket_sets = basket.drop('TransactionID', axis=1)

# Apply the Apriori algorithm
frequent_itemsets = apriori(basket_sets, min_support=0.0075, use_colnames=True)

frequent_itemsets
```

##### <a id="4-2-multi-itemsets"></a>**4.2 Multi Itemsets**

```python
multi_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) > 1)]
print(multi_itemsets)
```

##### <a id="4-3-association-rules"></a>**4.3 Association Rules**

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=.001)

# rules = rules[(rules['lift'] >= .01) & (rules['confidence'] >= 0.005)]

# round all values to 2 decimal places
rules = rules.round(2)
rules['antecedents'] = rules['antecedents'].apply(lambda x: list(x)[0]).astype("unicode")
rules['consequents'] = rules['consequents'].apply(lambda x: list(x)[0]).astype("unicode")

rules.head() 
```

##### <a id="4-4-support-vs-confidence"></a>**4.4 Support vs Confidence**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="support", y="confidence", size="lift", hue="lift", data=rules, palette="Blues")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Support vs Confidence")
plt.show()
```

##### <a id="4-5-lift-heatmap"></a>**4.5 Lift Heatmap**

```python
# Assuming 'rules' DataFrame is already defined
pivot = rules.pivot(index='antecedents', columns='consequents', values='lift')

# Generate heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, cmap="cividis")  
plt.title("Lift Heatmap")
plt.show()
```

##### <a id="4-6-itemset-network"></a>**4.6 Itemset Network**

```python
import networkx as nx
import matplotlib.pyplot as plt

# here we will create a new directed graph from edges
G = nx.from_pandas_edgelist(rules, source='antecedents', target='consequents', edge_attr=True, create_using=nx.DiGraph())

# new matplotlib figure and axes object
fig, ax = plt.subplots(figsize=(12, 8))  # Increased figure size

# Adjust node layout to reduce overlap
pos = nx.spring_layout(G, k=0.5)  # Adjust k to spread nodes out

# Draw the graph with improved visibility
nx.draw(G, pos, with_labels=True, node_color="#0057B7", node_size=2500, 
        font_size=12, font_weight="bold", edge_color="gray", width=1.5, ax=ax)

# Add edge labels with larger font size
labels = nx.get_edge_attributes(G, 'lift')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=10, ax=ax)

# Add title with improved styling
ax.set_title("Itemset Network", fontsize=16, fontweight="bold")

# Show the plot
plt.show()
plt.savefig("my_figure.png", dpi=300, bbox_inches='tight')
```

## <a id="6-recommendation-function"></a>**6. Recommendation Function**


**The Recommendation Function** uses the association rules generated from the **Apriori Algorithm** to recommend products to customers based on their previous purchases. The function takes an item that a user has already purchased (referred to as the **antecedent**) and finds the most likely items they would purchase next (the **consequents**).

**Objective:**
The goal of this function is to provide personalized product recommendations to customers. By leveraging the insights from the association rules, we can suggest items that are commonly bought together with a given product, enhancing the shopping experience and promoting cross-selling or upselling.

**Steps in the Recommendation Process:**

**Identify the Antecedent Item:**

The antecedent is the item that the customer has already purchased or is interested in. For example, if the customer buys **whole milk**, we want to recommend other items that are often bought with milk.

**Find Association Rules:**

The association rules generated earlier in the project contain **antecedents** (items that were purchased first) and **consequents** (items that are likely to be bought next). The function searches the association rules for any that have the given antecedent item.

**Sort by Confidence or Lift:**

The rules are then sorted based on **confidence** or **lift**, which measure the strength of the association between the antecedent and consequent. Confidence represents how likely the consequent is to be purchased when the antecedent is purchased, while lift shows the likelihood of the consequent being bought together with the antecedent relative to how often it is purchased overall.

**Top N Recommendations:**

The function selects the top N recommendations based on the sorted confidence or lift values. These are the most relevant and likely items to be recommended to the user.

```python
def recommend_item(antecedent_item, rules, top_n=5):
    """
    Recommend items based on a specific antecedent item.

    Parameters:
    antecedent_item (str): The item that the user has purchased or is interested in.
    rules (DataFrame): The association rules DataFrame containing antecedents, consequents, confidence, and lift.
    top_n (int): The number of top recommendations to return.

    Returns:
    DataFrame: A DataFrame containing the top recommended items based on the antecedent.
    """
    # Filter the rules where the antecedent matches the given item
    recommendations = rules[rules['antecedents'].apply(lambda x: antecedent_item in x)]
    
    # Sort by confidence and select top_n recommendations
    recommendations_sorted = recommendations[['antecedents', 'consequents', 'confidence', 'lift']].sort_values(by='confidence', ascending=False)
    
    # Limit to the top_n recommendations
    top_recommendations = recommendations_sorted.head(5)

    return top_recommendations

# Example: Get recommendations for "whole milk"
print(recommend_item("whole milk", rules))
```
